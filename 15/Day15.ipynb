```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.start = None
    
    def append_node(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
        else:
            temp = self.start
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
    
    def print_list(self):
        current = self.start
        while current is not None:
            print(current.data, end=" ")
            current = current.next
        print()
    
    def reverse_list(self):
        current = self.start
        arr = []
        initial = self.start
        while current is not None:
            arr.append(current.data)
            current = current.next
        arr = arr[::-1]
        i = 0
        while initial is not None:
            initial.data = arr[i]
            initial = initial.next
            i += 1

    def get_length(self):
        current = self.start
        length = 0
        while current is not None:
            length += 1
            current = current.next
        return length

    def delete_middle(self):
        length = self.get_length()
        current = self.start
        counter = length // 2
        while counter > 1:
            current = current.next
            counter -= 1
        current.next = current.next.next
    
    def del_middle_element(self):
        slow_ptr = self.start
        fast_ptr = self.start
        prev_ptr = None

        while fast_ptr is not None and fast_ptr.next is not None:
            fast_ptr = fast_ptr.next.next
            prev_ptr = slow_ptr
            slow_ptr = slow_ptr.next

        prev_ptr.next = slow_ptr.next
    def delete_node(self, target_data):
        current = self.start
        prev = None
    
        while current is not None and current.data != target_data:
            prev = current
            current = current.next
    
        if current is None:
            print(f"Node with data {target_data} not found.")
            return
    
        if prev is None:
            self.start = current.next
        else:
            prev.next = current.next
    
        del current
    def delete_node_after_and_before(self,n):
        current=self.start
        if current.data==n:
            print("before does not exist")
            current=current.next.next
        else:
            while current.next.next is not None and current.next.next.data!=n :
                current=current.next
            current.next=current.next.next
            current=current.next
            current.next=current.next.next
    def insert_after(self, after_value, val_to_be_inserted):
        current = self.start
    
        while current is not None and current.data != after_value:
            current = current.next
    
        if current is None:
            print(f"After Value {after_value} does not exist.")
            return
    
        new_node = Node(val_to_be_inserted)
        new_node.next = current.next
        current.next = new_node


            
obj = LinkedList()

for i in range(12):
    obj.append_node(i)

obj.print_list()
obj.reverse_list()
obj.print_list()
obj.delete_middle()
obj.print_list()
obj.delete_node(2)
obj.print_list()
obj.delete_node_after_and_before(4)
obj.delete_middle()
obj.print_list()
obj.get_length()
obj.print_list()
obj.insert_after(9,15)
obj.print_list()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def merge_and_sort_linked_lists(list1, list2):
    dummy = Node(0)
    current = dummy

    while list1 is not None and list2 is not None:
        if list1.value < list2.value:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    if list1 is not None:
        current.next = list1
    else:
        current.next = list2

    return dummy.next

def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Define linked lists
a1 = Node(18)
a1.next = Node(28)
a1.next.next = Node(58)
a1.next.next.next = Node(119)
a1.next.next.next.next = Node(120)

a2 = Node(17)
a2.next = Node(19)
a2.next.next = Node(57)
a2.next.next.next = Node(68)
a2.next.next.next.next = Node(123)
a2.next.next.next.next.next = Node(567)

# Merge and sort linked lists
sorted_list = merge_and_sort_linked_lists(a1, a2)

# Print sorted linked list
print_linked_list(sorted_list)
```
